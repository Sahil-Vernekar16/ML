```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import difflib
```

```python
movies = pd.read_csv('movies.csv')
```

```python
movies.head()
```

```python
features = ['genres', 'keywords' , 'title', 'cast' , 'director']
```

```python
for feature in features:
    movies[feature] = movies[feature].fillna('')
```

```python
combined_features = movies['genres'] + ' ' + movies['keywords'] + ' ' + movies['title'] + ' ' + movies['cast'] + ' ' + movies['director']
```

```python
vectorizer = TfidfVectorizer()
```

```python
combined_features = vectorizer.fit_transform(combined_features)
```

```python
similarity = cosine_similarity(combined_features)
```

```python
print(similarity.shape)
```

```python
user_input = input('Enter a movie:')
```

```python
all_movies = movies['title'].tolist()
```

```python
find_similar_movies = difflib.get_close_matches(user_input, all_movies)
```

```python
find_similar_movies
```

```python
index_of_movie = movies[movies.title == find_similar_movies[0]]['index'].values[0]
```

```python
index_of_movie
```

```python
similar_movies = list(enumerate(similarity[index_of_movie]))
```

```python
similar_movies = sorted(similar_movies, key= lambda x:x[1], reverse=True)
```

```python
similar_movies
```

```python
for i in range(30):
  index = similar_movies[i][0]
  title = movies[movies.index == index]['title'].values[0]
  print(i+1 , title)
```

# Movie Recommendation System

```python
user_input = input('Enter a movie:')
find_similar_movies = difflib.get_close_matches(user_input, all_movies)
index_of_movie = movies[movies.title == find_similar_movies[0]]['index'].values[0]
similar_movies = list(enumerate(similarity[index_of_movie]))
similar_movies = sorted(similar_movies, key= lambda x:x[1], reverse=True)
for i in range(30):
  index = similar_movies[i][0]
  title = movies[movies.index == index]['title'].values[0]
  print(i+1 , title)
```
