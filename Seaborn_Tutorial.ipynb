```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
tips  = sns.load_dataset('tips')
```

```python
tips.head()
```

# RelPlot

```python
sns.set_theme()
```

```python
sns.relplot(x='total_bill', y='tip', data=tips,col='time',hue='smoker',style='smoker',size='size')
```

# Distribution Plot

```python
from sklearn.datasets import fetch_california_housing
```

```python
housing_dataset = fetch_california_housing()
```

```python
housing_df = pd.DataFrame(housing_dataset.data,columns=housing_dataset.feature_names)
```

```python
housing_df.head()
```

```python
housing_df['Price'] = housing_dataset.target
```

```python
housing_df.head()
```

```python
sns.distplot(housing_df['Price'])
```

# Correlation

```python
plt.figure(figsize=(10,10))
sns.heatmap(housing_df.corr(), annot=True, cmap='Blues',cbar=True, square=True)
```
