```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme()
```

```python
plt.quiver(0,0,4,5, scale_units='xy', angles='xy', scale=1,color='g')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

# Vector Addition

```python
vector1 = np.array([0,0,2,3])
vector2 = np.array([0,0,3,-2])
result = vector1 + vector2
print(result)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0,0,3,-2, scale_units='xy', angles='xy', scale=1,color='g')
plt.quiver(0,0,5,1, scale_units='xy', angles='xy', scale=1, color='y')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

# Vector Subtraction

```python
vector1 = np.array([0,0,2,3])
vector2 = np.array([0,0,3,-2])
result = vector1 - vector2
print(result)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0,0,3,-2, scale_units='xy', angles='xy', scale=1,color='g')
plt.quiver(0,0,-1,5, scale_units='xy', angles='xy', scale=1, color='y')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

# Multiply a Vector by a Scalar

```python
vector1 = np.array([0,0,2,3])
scalar = 2
result = vector1 * scalar
print(result)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0,0,4,6, scale_units='xy', angles='xy', scale=1,color='y')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

# Dot Product

```python
a = np.array([2,3])
b = np.array([4,4])
result = np.dot(a,b)
print(result)
```

# Cross Product

```python
c = np.array([5,10,20])
d= np.array([18,32,50])
result = np.cross(c,d)
print(result)
```

# Projection of a on v

```python
a = np.array([23,45,62])
v = np.array([45,82,67])

magnitude_of_v = np.sqrt(sum(v**2))
result = (np.dot(a,v)/magnitude_of_v**2) * v
print(result)
```
