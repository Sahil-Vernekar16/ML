```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# Data Collection and Preprocessing

```python
dataset_df = pd.read_csv('train.csv')
```

```python
dataset_df.head()
```

```python
dataset_df.isnull().sum()
```

```python
dataset_df['Age'] = dataset_df['Age'].fillna(dataset_df['Age'].mean())
```

```python
dataset_df = dataset_df.drop(columns='Cabin', axis=1)
```

```python
print(dataset_df['Embarked'].mode()[0])
```

```python
dataset_df['Embarked'] = dataset_df['Embarked'].fillna(dataset_df['Embarked'].mode()[0])
```

```python
dataset_df.head()
```

```python
encoder = LabelEncoder()
```

```python
dataset_df['Sex'] = encoder.fit_transform(dataset_df['Sex'])
dataset_df['Embarked'] = encoder.fit_transform(dataset_df['Embarked'])
```

```python
X = dataset_df.drop(columns=['PassengerId', 'Name', 'Ticket', 'Survived'], axis=1)
Y = dataset_df['Survived']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Model Building and Testing

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
training_prediction = model.predict(X_train)
training_accuracy = accuracy_score(Y_train, training_prediction)
print(training_accuracy)
```

```python
test_prediction = model.predict(X_test)
test_accuracy = accuracy_score(Y_test, test_prediction)
print(test_accuracy)
```
