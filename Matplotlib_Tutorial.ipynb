```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.linspace(0, 10, 100)
b = np.sin(a)
c = np.cos(a)
print(a)
print(b)
print(c)
```

```python
#Sin Wave
plt.plot(a,b)
plt.show()
```

```python
#Cos Wave
plt.plot(a,c , 'g*')
plt.xlabel('Angle')
plt.ylabel('Cos Value')
plt.title('Cos Wave')
plt.show()
```

# Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
ax.bar(langs,students)
plt.show()
```

# Pie Chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
ax.pie(students,labels=langs , autopct = '%1.1f%%', shadow = True)
plt.show()
```

# Parabola

```python
x = np.linspace(-10,10,100)
y = x**2
plt.plot(x,y)
plt.show()
```

# Scatter Plot

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
plt.scatter(x,y , color='r')
plt.scatter(x,z, color='g')
plt.show()
```

# 3D Scatter Plot

```python
fig2 = plt.figure()
ax = plt.axes(projection='3d')
x = 20 * np.random.random(100)
y = np.sin(x)
z = np.cos(x)
ax.scatter(y,z,x,c=x)
plt.show()
```
