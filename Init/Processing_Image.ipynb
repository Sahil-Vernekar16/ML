# PILLOW

```python
from PIL import Image
```

```python
dog1 = Image.open(r"dog.jpg")
```

```python
dog1_reshaped = dog1.resize((200,200))
```

```python
dog1_reshaped = dog1_reshaped.save("dog1_reshaped.jpg")
```

```python
dog1 = mpimg.imread('dog1_reshaped.jpg')
plt.imshow(dog1)
plt.show()
```

# OPEN CV

```python
import cv2
```

```python
dog2 = cv2.imread('dog.jpg')
```

```python
dog2_grayscale = cv2.cvtColor(dog2,cv2.COLOR_RGB2GRAY)
```

```python
from google.colab.patches import cv2_imshow
```

```python
cv2_imshow(dog2_grayscale)
```
