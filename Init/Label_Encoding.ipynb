```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

# Dataset 1

```python
dataset_df = pd.read_csv('data.csv')
```

```python
dataset_df.head()
```

```python
label_encoder = LabelEncoder()
```

```python
labels = label_encoder.fit_transform(dataset_df['diagnosis'])
```

```python
dataset_df['target'] = labels
```

```python
dataset_df.head()
```

```python
dataset_df.tail()
```

```python
dataset_df.value_counts('diagnosis')
```

```python
dataset_df.value_counts('target')
```
