```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

# Data Collection and Preprocessing

```python
dataset_df = pd.read_csv('train.csv')
```

```python
dataset_df.head()
```

```python
dataset_df.isnull().sum()
```

```python
X = dataset_df.drop('quality',axis=1)
```

```python
X
```

```python
Y = dataset_df['quality'].apply(lambda  y: 1 if y>=7 else 0)
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y,  random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Model Building and Testing

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train, Y_train)
```

```python
training_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(training_prediction, Y_train)
print(training_data_accuracy)
```

```python
testing_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(testing_prediction, Y_test)
print(testing_data_accuracy)
```

# Custom Input

```python
input = (7.3,0.65,0.0,1.2,0.065,15.0,21.0,0.9946,3.39,0.47,10.0)
input = np.asarray(input)
input_reshaped = input.reshape(1,-1)
prediction = model.predict(input_reshaped)
print(prediction)
```
